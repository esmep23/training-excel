```python
import pandas as pd
import numpy as np
import random

# Definimos listas de datos base para generar la suciedad
nombres_base = ["Juan", "Marta", "Carlos", "Ana Maria", "Elena", "Pedro", "LucÃ­a", "Diego", "Sofia", "Roberto"]
apellidos_base = ["Perez", "Gomez", "Ruiz", "Lopez", "Saenz", "Garcia", "Fernandez", "Torres"]
regiones_base = ["EspaÃ±a", "ES", "esp", "MÃ©xico", "mex", "MEXICO", "Argentina", "ARG", "arg"]

# Lista para almacenar los 100 registros
registros = []

for _ in range(100):
    # 1. Crear Nombre sucio (MayÃºsculas locas y espacios)
    nom = random.choice(nombres_base) + " " + random.choice(apellidos_base)
    estilo = random.choice(['upper', 'lower', 'proper', 'messy'])
    if estilo == 'upper': nom = nom.upper()
    elif estilo == 'lower': nom = nom.lower()

    # AÃ±adimos espacios aleatorios al inicio y final
    nombre_sucio = " " * random.randint(1, 4) + nom + " " * random.randint(1, 4)

    # 2. Crear ID_Producto inconsistente (ej. PROD-001-X, prod--002, X-PROD-004)
    num = str(random.randint(100, 999))
    formatos_id = [f"PROD-{num}-X", f"prod--{num}", f"  PROD{num}-Y", f"X-PROD-{num}", f"PR-{num}-99"]
    id_sucio = random.choice(formatos_id)

    # 3. Elegir una RegiÃ³n inconsistente
    region_sucio = random.choice(regiones_base)

    registros.append([nombre_sucio, id_sucio, region_sucio])

# Creamos el DataFrame
df_100 = pd.DataFrame(registros, columns=['Nombre_Cliente', 'ID_Producto', 'Region'])

# Guardamos el archivo
nombre_archivo = 'ejercicio_100_leads_sucios.csv'
df_100.to_csv(nombre_archivo, index=False, encoding='utf-8-sig')

print(f"âœ… Â¡Ã‰xito! Se ha generado el dataset con {len(df_100)} registros.")
print(f"ðŸ‘‰ Descarga el archivo '{nombre_archivo}' desde el icono de carpeta a la izquierda.")
```
